```python
class film: 
    def __init__(self, name, year, country, janr, cost):

        self.name = name
        self.year = year
        self.country = country
        self.janr = janr
        self.cost = cost


    def get_name(self):
        return self.name

    def get_country(self):
        return self.country

    def get_janr(self):
        return self.janr

    def get_year(self):
        return self.year

    
    def get_cost(self):
        return self.cost

    def get_info(self):
        print(self.get_name())
        print(self.get_country())
        print(self.get_janr())
        print(self.get_year())
        print(self.get_cost())

    def set_name(self, name):
        self.name = name
    
    def set_year(self, year):
        self.year = year
    
    def set_country(self, country):
        self.country = country

    def set_janr(self, janr):
        self.janr = janr

    def set_cost(self, cost):
        self.cost = cost

    def redact(self, inf):
        redact = input()
        match redact:
            case "name":
                self.name = inf
            case "year":
                self.year = inf
            case "country":
                self.country = inf
            case "janr":
                self.janr = inf
            case "cost":
                self.cost = inf

        return self.get_info()





class film_tutor:
    def __init__(self):
        self.films = []

    def add(self, film):
        self.films.append(film)

    def oldest_film(self):
        for film in self.films:
            min = 1000000
            if film.get_year() < min:
                min = film.get_year()
            
        for film in self.films:
            if film.get_year() == min:
                film.get_info()


    def get_average(self):
        value = 0
        for film in self.films:
            value += film.get_cost()

        value = value / len(self.films)
        
        for film in self.films:
            if film.get_cost() > value:
                print("Фильмы, прокат выше среднего")
                print(film.get_name())

    def alphabet_sort(self):
        sorted_films = []
        for film in self.films:
            sorted_films.append(film.get_name())
            
        return(sorted(sorted_films))

    def name_search(self, name):
        for film in self.films:
            if name == film.get_name():
                return film.get_info()

    


        

    

film_1 = film("Good Fellas", 1997, "USA", "Drama", 2000000)
film_2 = film("Aviator", 2007, "USA", "Bio", 10000000)
film_3 = film("Bronx Tale", 1985, "USA", "Drama", 5000000)
film_tutor_1 = film_tutor()
film_tutor_1.add(film_1)
film_tutor_1.add(film_2)
film_tutor_1.add(film_3)


print(film_tutor_1.oldest_film())
print(film_tutor_1.alphabet_sort())
print(film_tutor_1.name_search("Aviator"))
print(film_1.redact("Italia"))
```
